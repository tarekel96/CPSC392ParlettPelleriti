```python
penguin_url = 'https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/penguins.csv'
titanic_url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv'
penguin_DF = pd.read_csv(penguin_url)
titanic_DF = pd.read_csv(titanic_url)
```

```python
penguin_DF['bill_ratio'] = penguin_DF['bill_length_mm'] / penguin_DF['bill_depth_mm']
print(penguin_DF.columns)
```

```python
lil_boys = penguin_DF.loc[penguin_DF['body_mass_g'] < 4000]
lil_boys.head()
```

```python
Adelie = penguin_DF['species'] == 'Adelie'
DF_Adelie = penguin_DF.loc[Adelie]
bl_Adelie = np.mean(DF_Adelie['bill_length_mm'])

Chinstrap_DF = penguin_DF.loc[penguin_DF['species'] == 'Chinstrap']
bl_Chinstrap = np.mean(Chinstrap_DF['bill_length_mm'])

Gentoo = penguin_DF.loc[penguin_DF['species'] == 'Gentoo']
bl_Gentoo = np.mean(Gentoo['bill_length_mm'])

print(bl_Adelie)
print(bl_Chinstrap)
print(bl_Gentoo)
```
